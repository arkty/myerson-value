```python
from ipywidgets import widgets
from IPython.display import display
from io import StringIO
from numpy.linalg import matrix_power
from numpy import matrix
from numpy import loadtxt
def s(a_k, i, k):
    n = len(a_k[1])
    prefix = sum(a_k[k][i, j] for j in range(n))
    mainSum = sum(
        sum(
            a_k[r][l, i] * sum(
                a_k[k - r][i, j]
                for j in range(n)
            )
            for r in range(1, k)
        ) + a_k[k][l, i]
        for l in range(n)
    )
    return prefix + mainSum
    
def myerson(a, k):
    a_k = [matrix_power(a, i) for i in range(k + 1)]
    return [s(a_k, i, k) for i in range(len(a))]

def on_calculate(c):
    I = loadtxt(StringIO(matrixInput.value))
    output.value = ", ".join(map(str, myerson(I, int(kInput.value))))
    
matrixInput = widgets.Textarea(placeholder='Введите матрицу инцидентности')
kInput = widgets.Text(placeholder='Введите k')
calculate = widgets.Button(description='Найти вектор')
calculate.on_click(on_calculate)

output = widgets.Label(value="") 
display(matrixInput)
display(kInput)
display(calculate)
display(output)
```
